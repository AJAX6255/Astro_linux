```python
%matplotlib inline
import numpy as np
from gatspy.periodic import LombScargleFast, LombScargle # Time series analysis
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd
import seaborn # Additional plotting
# Set aesthetic parameters in one step.
seaborn.set()
```

```python
# Data from AAVSO
star = "LX_Cyg"

# Load the data from csv file
light_curve = pd.read_csv("LX_Cyg.csv", usecols=["JD", "Magnitude"])
# Remove "<" symbol from Magnitude data
light_curve.Magnitude = [m.replace("<", "") for m in light_curve.Magnitude]
light_curve.Magnitude = pd.to_numeric(light_curve.Magnitude)

# Creating numpy array from dataframe
mag = np.array([x for x in light_curve.Magnitude])
time = np.array([x for x in light_curve.JD])
```

```python
# Plot the raw data.
plt.scatter(time, mag)
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Magnitude")
plt.title("Raw light curve of " +star)
plt.show()
```

The $\textbf{Lomb-Scargle periodogram}$ is an algorithm used for detecting and characterizing periodic signals in unevenly-sampled data. It efficiently computes a Fourier-like power spectrum estimator from the data, resulting in an intuitive means of determining the period of oscillation.

$\textbf{Classical}$: 
The power spectrum for evenly spaced data is given by $P_g = |\mathcal{F} \{ g\}|^2 = \bigg( \big( \sum_n g_n \cos(2 \pi f t_n) \big)^2 + \big( \sum_n g_n \sin(2 \pi f t_n) \big)^2 \bigg) $

$\textbf{Lomb-Scarge}$: $P_{LS} = \frac{1}{2} \bigg\{ \frac{\bigg( \sum_n g_n \cos(2 \pi f [t_n - \tau]) \bigg)^2}{\sum_n \cos ^2(2 \pi f [t_n - \tau] } + \frac{\bigg( \sum_n g_n \sin(2 \pi f [t_n - \tau]) \bigg)^2}{\sum_n \sin^2(2 \pi f [t_n - \tau])} \bigg\}$





```python
# Calculate the Lomb-Scarge Periodogram and find the most significant period.
ls = LombScargleFast() #LombScargle() for more exact
ls.optimizer.period_range = (1, 2000)
ls.fit(time, mag)
period = ls.best_period
print("Best period: " + str(period) + " days")
```

```python
# The Periodogramm
periods = np.linspace(1,2000,5000)
ls.score(periods)
scores = ls.score(periods)
plt.plot(periods,scores)
plt.xlabel("Period")
plt.ylabel("Power")
plt.xlim([0,2000])
plt.title("Lomb Scargle Periodogram of " +star)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.stats import LombScargle

t = time
y = mag

frequency, power = LombScargle(t, y, nterms=2).autopower()
best_frequency = frequency[np.argmax(power)]
print(best_frequency)
phase_fit = np.linspace(0, 1)
y_fit = LombScargle(t, y, nterms=2).model(phase_fit / best_frequency,
                                    frequency=best_frequency)
phase = (t * best_frequency) % 1

fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(phase_fit, y_fit, color='black', zorder=10)
ax.errorbar(phase, y, fmt='o', mew=0, capsize=0, elinewidth=1.5)
ax.plot(phase_fit, y_fit, color='black')
ax.invert_yaxis()
ax.set(xlabel='phase',
       ylabel='magnitude',
       title='phased data at frequency={0:.5f}'.format(best_frequency))
plt.show()

print(y_fit)

# fig, ax = plt.subplots(figsize=(10, 7))
# plt.plot(phase_fit, y_fit)
# plt.plot(phase, y)
# plt.plot(phase_fit, y_fit, color='black')
# plt.gca().invert_yaxis()
# plt.xlabel=('phase')
# plt.ylabel=('magnitude')
# plt.title=('phased data at frequency={0:.5f}'.format(best_frequency))
# plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.stats import LombScargle

rand = np.random.RandomState(42)
t = 100 * rand.rand(100)
dy = 0.1
y = np.sin(2 * np.pi * t) + dy * rand.randn(100)

frequency, power = LombScargle(t, y, dy).autopower(minimum_frequency=0.1,
                                                   maximum_frequency=1.9)
best_frequency = frequency[np.argmax(power)]
phase_fit = np.linspace(0, 1)
y_fit = LombScargle(t, y, dy).model(t=phase_fit / best_frequency,
                                    frequency=best_frequency)
phase = (t * best_frequency) % 1

fig, ax = plt.subplots(figsize=(6, 4.5))
ax.errorbar(phase, y, dy, fmt='o', mew=0, capsize=0, elinewidth=1.5)
ax.plot(phase_fit, y_fit, color='black')
ax.invert_yaxis()
ax.set(xlabel='phase',
       ylabel='magnitude',
       title='phased data at frequency={0:.2f}'.format(best_frequency))
```
