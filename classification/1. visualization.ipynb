### Feature vs Feature
Due to the large number of features, it is difficult to visualize the data.  Three solutions to this problem:
* Dimensionality reduction with PCA to 2 or 3 dimensions.


* Dimensionality reduction with T-SNE to 2 or 3 dimensions. This method has the added benefit of preserving the clustering in the original data set.


* Plot all possible 2D or 3D graphs of the features, and manually look for clustering. This method has the added benefit of showing which feature combinations produce good clustering. Since 64 features are used in the data set, there will be a total of 2016 2D graphs or 41664 3D graphs. I will be using just the 2D graphs because there is an unreasonable amount of 3D graphs to sieve through.

### Dependencies

```python
%matplotlib inline
import sqlite3 # Database library.
import os # Folder management library.
import pickle # Serializing module.
import numpy as np # Scientific computing library.
from sklearn.manifold import TSNE # T-SNE library.
import matplotlib.pyplot as plt # Plotting library.
```

### Import Data from Database

```python
# Initializing database and cursor.
star_data_db = sqlite3.connect('star_data.db')
star_data_cursor = star_data_db.cursor()

# Retrieving star_data from database.
classes = ['cep_1o', 'cep_f', 'dsct','eb_ec', 'eb_ed', 'eb_esd', 'lpv_mira_agb_c', 'lpv_mira_agb_o', 'lpv_osarg_agb',
           'lpv_osarg_rgb', 'lpv_srv_agb_c', 'lpv_srv_agb_o', 'rrab', 'rrc', 'rrd', 'rre', 't2cep']

X, Y = [], []
for label, classv in enumerate(classes):
    temp_X, temp_Y = [], []
    star_data_cursor.execute('SELECT star_features FROM '+classv)
    for row in star_data_cursor.fetchall()[:600]:
        # Deserializing features.
        features = pickle.loads(row[0])
        temp_X.append(features)
        temp_Y.append([label])
    X.append(np.array(temp_X))
    Y.append(np.array(temp_Y))
        

# Close cursor and database    
star_data_cursor.close
star_data_db.close()

print('Done.')
```

### T-SNE

```python
# Flatten lists.
X = [item for sublist in X for item in sublist]
Y = [item for sublist in Y for item in sublist]
# Make lists numpy arrays.
X = np.array(X)
Y = np.array(Y)
print(X.shape)

X_embedded = TSNE(n_components=2).fit_transform(X)
print(X_embedded.shape)
```

### T-SNE Plot

```python
# All 17 star types as different colours.
plt.scatter(X_embedded[:,0], X_embedded[:, 1], c=Y)
plt.title('All 17 Stars as Different Colours')
plt.colorbar()
plt.show()

# Only t2ceps are a different colour.
Y = Y.tolist()
Y = [item for sublist in Y for item in sublist]
new_Y = []
for n in Y:
    if n !=16:
        new_Y.append(0)
    if n==16:
        new_Y.append(1)
        
plt.scatter(X_embedded[:,0], X_embedded[:, 1], c=new_Y)
plt.title('T2ceps vs All')
plt.colorbar()
plt.show()
```
