### Dependencies

```python
import sqlite3 # Database library.
import os # Folder management library.
import pickle # Serializing module.
import numpy as np # Scientific computing library.
import matplotlib.pyplot as plt # Plotting library.
from mpl_toolkits.mplot3d import Axes3D # 3D plotting tool.
from sklearn.neighbors import LocalOutlierFactor # Outlier Dection library.
```

### Import Data from Database

```python
# Initializing database and cursor.
star_data_db = sqlite3.connect('star_data.db')
star_data_cursor = star_data_db.cursor()

# Retrieving star_data from database.
classes = ['cep_1o', 'cep_f', 'dsct','eb_ec', 'eb_ed', 'eb_esd', 'lpv_mira_agb_c', 'lpv_mira_agb_o', 
           'lpv_osarg_agb','lpv_osarg_rgb', 'lpv_srv_agb_c', 'lpv_srv_agb_o', 'rrab', 'rrc', 'rrd', 
           'rre', 't2cep']

X, Y = [], []
for label, classv in enumerate(classes):
    temp_X, temp_Y = [], []
    star_data_cursor.execute('SELECT star_features FROM '+classv)
    for row in star_data_cursor.fetchall()[:600]:
        # Deserializing features.
        features = pickle.loads(row[0])
        temp_X.append(features)
        temp_Y.append([label])
    X.append(np.array(temp_X))
    Y.append(np.array(temp_Y))
    
names = []
for classv in classes:
    temp_names = []
    star_data_cursor.execute('SELECT star_name FROM '+classv)
    for row in star_data_cursor.fetchall()[:600]:
        temp_names.append(row[0])
    names.append(temp_names)
        

# Close cursor and database    
star_data_cursor.close
star_data_db.close()

print('done')
```

### Visualization of Local Outlier Factor on T2cep Data
Each star is described using 64 features, ergo, the data is 64-dimensional.  Since the data cannot be visualized, a 2D sample data set of t2ceps is used to visually understand the outlier detection algorithm -- Local Outlier Factor. The two features used to compute the sample data set are 'StetsonK' and 'Rcs'. More info on the features can be found on http://isadoranun.github.io/tsfeat/FeaturesDocumentation.html.

<img src="img/StetsonKvsRcs.png", width=600, height=auto>

```python
%matplotlib inline

feature_names = ['Amplitude', 'AndersonDarling', 'Autocor_length', 'Beyond1Std', 'CAR_mean', 
                 'CAR_sigma', 'CAR_tau', 'Con', 'Eta_e', 'FluxPercentileRatioMid20', 
                 'FluxPercentileRatioMid35', 'FluxPercentileRatioMid50','FluxPercentileRatioMid65',
                 'FluxPercentileRatioMid80', 'Freq1_harmonics_amplitude_0', 
                 'Freq1_harmonics_amplitude_1', 'Freq1_harmonics_amplitude_2', 
                 'Freq1_harmonics_amplitude_3', 'Freq1_harmonics_rel_phase_0',
                 'Freq1_harmonics_rel_phase_1', 'Freq1_harmonics_rel_phase_2', 
                 'Freq1_harmonics_rel_phase_3', 'Freq2_harmonics_amplitude_0', 
                 'Freq2_harmonics_amplitude_1', 'Freq2_harmonics_amplitude_2', 
                 'Freq2_harmonics_amplitude_3', 'Freq2_harmonics_rel_phase_0',
                 'Freq2_harmonics_rel_phase_1', 'Freq2_harmonics_rel_phase_2', 
                 'Freq2_harmonics_rel_phase_3', 'Freq3_harmonics_amplitude_0', 
                 'Freq3_harmonics_amplitude_1', 'Freq3_harmonics_amplitude_2', 
                 'Freq3_harmonics_amplitude_3', 'Freq3_harmonics_rel_phase_0', 
                 'Freq3_harmonics_rel_phase_1','Freq3_harmonics_rel_phase_2', 
                 'Freq3_harmonics_rel_phase_3', 'Gskew', 'LinearTrend', 'MaxSlope','Mean', 
                 'Meanvariance', 'MedianAbsDev', 'MedianBRP', 'PairSlopeTrend', 'PercentAmplitude', 
                 'PercentDifferenceFluxPercentile', 'Period_fit', 'PeriodLS', 'Psi_CS', 'Psi_eta', 'Q31',
                 'Rcs', 'Skew', 'SlottedA_length', 'SmallKurtosis', 'Std', 'StetsonK', 'StetsonK_AC', 
                 'StructureFunction_index_21', 'StructureFunction_index_31', 'StructureFunction_index_32',
                 'Colour']

# A 2D sample dataset of t2ceps.
x_temp = X[16][:,feature_names.index('StetsonK')]
y_temp = X[16][:,feature_names.index('Rcs')]
sample_data = np.column_stack((x_temp, y_temp))


# Fitting the Local Outlier Detection model to the sample data set. 
lof = LocalOutlierFactor(n_neighbors=20, contamination=10/len(sample_data))
outlier_pred = lof.fit_predict(sample_data)

print('1: Data point predicted to not be an outlier. \n',
      '-1: Data pont predicted to be an outlier.')
print(outlier_pred)
print('the number of t2ceps is ', len(outlier_pred), '\n')

# Find the indices if all outliers.
outlier_ind = np.where(outlier_pred == -1)[0]
print('The indices of the outliers are:')
print(outlier_ind)
print(" The num of outliers is ", len(outlier_ind), '\n')

# Find the names of the outliers, and their coordinates.
outlier_x, outlier_y = [], []
print('The outliers are:')
for ind in outlier_ind:
    print(names[16][ind])
    outlier_x.append(sample_data[ind][0])
    outlier_y.append(sample_data[ind][1])

# Plot outliers and inliers.
plt.figure(figsize=(9,6))
plt.scatter(sample_data[:,0], sample_data[:,1] , marker='o', c='white', label='normal observation')
plt.scatter(outlier_x, outlier_y , marker='o', c='red', label='abnormal observation')
plt.xlabel('StetsonK', size='medium', family = 'monospace')
plt.ylabel('Rcs', size='medium', family = 'monospace')
plt.legend(loc='upper left')
plt.show()
plt.close()
```

### 3D plot: StetsonK vs Rcs vs Mean Magnitude
The features of fainter/lower-mean-magnitude stars are less defined as their measurments are much noisier.  Hence, an easy way to check if an outlier is the consequence of measuring a fainter star is to plot the mean magnitude of the star on a third dimension.

```python
%matplotlib notebook

# Plot features.
x_plt = 'StetsonK'
y_plt = 'Rcs'
z_plt = 'Mean'

x = X[16][:,feature_names.index(x_plt)].tolist()
y = X[16][:,feature_names.index(y_plt)].tolist()
z = X[16][:,feature_names.index(z_plt)].tolist()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z, c='r', marker='o')
ax.set_xlabel(x_plt)
ax.set_ylabel(y_plt)
ax.set_zlabel(z_plt+' Magnitude')
plt.show()
```

### 10 Best Outlier Candidates

```python
%matplotlib inline

# Reconstructing data for readibility.
names_t2cep = names[16]
x_temp = X[16][:,feature_names.index('StetsonK')]
y_temp = X[16][:,feature_names.index('Rcs')]
sample_data = np.column_stack((x_temp, y_temp))

best_outliers = []

# Recursively applying lof algorithm to obtain 10 best outliers.
for n in range(0,10):
    
    # Fitting the Local Outlier Detection model to the sample data set, and looking for 1 outlier. 
    lof = LocalOutlierFactor(n_neighbors=20, contamination=1/len(sample_data))
    outlier_pred = lof.fit_predict(sample_data)

    # Find the indices if all outliers.
    outlier_ind = np.where(outlier_pred == -1)[0]

    # Find the names of the outliers, and their coordinates.
    outlier_x, outlier_y = [], []
    print('The outliers are:')
    for ind in outlier_ind:
        print(names_t2cep[ind])
        best_outliers.append(names_t2cep[ind])
        outlier_x.append(sample_data[ind][0])
        outlier_y.append(sample_data[ind][1])

    # Plot outliers and inliers.
    plt.figure(figsize=(9,6))
    plt.scatter(sample_data[:,0], sample_data[:,1] , marker='o', c='white', label='normal observation')
    plt.scatter(outlier_x, outlier_y , marker='o', c='red', label='abnormal observation')
    plt.xlabel('StetsonK', size='medium', family = 'monospace')
    plt.ylabel('Rcs', size='medium', family = 'monospace')
    plt.legend(loc='upper left')
    plt.show()
    plt.close()
    
    # Remove outlier from data set.
    sample_data = np.delete(sample_data, outlier_ind[0], 0)
    names_t2cep = np.delete(names_t2cep, outlier_ind[0], 0)
    
print('The 10 best outliers are: ', best_outliers)

```
